```python
from sklearn.model_selection import train_test_split
import pandas as pd
from afinn import Afinn
from nltk.corpus import stopwords
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

dataset = pd.read_csv("sentiment_analysis.csv",encoding = 'utf-8')
dataset.head()
```

```python
sentimential_list = {'행복' : 0, '기쁨': 0, '분노' : 1, '불안' : 1, '당황' : 1, '슬픔' : 2, '상처' : 2}
dataset['label'] = dataset['sentiment_big'].map(sentimential_list)
dataset.head()
```

```python
import re

dataset['text'] = dataset['person1'].map(str) + dataset['system1'].map(str) + dataset['person2'].map(str)+ dataset['system2'].map(str) + dataset['person3'].map(str) + dataset['system3'].map(str)
new_data = {"text" : "ㅎㅎ","label": 0}
dataset = dataset.append(new_data, ignore_index = True)

x_data = dataset['text']
y_data = dataset['label'].values
```

```python
x_data = x_data.apply(lambda x : re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣ\\s]', " ", x)).values
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, stratify = y_data, test_size = 0.2, random_state = 42)
```

```python
data = pd.DataFrame()
data['train'] = x_train
text_list = data['train'].values
```

```python
from konlpy.tag import Okt
okt = Okt()

pos_texts = []
main_list = ["Noun","Verb"]
for t in range(len(text_list)) :
    a= []
    tokens = okt.pos(text_list[t], norm=True, stem=True)
    for token in tokens :
        if token[1] in main_list:
            a.append(token)
    pos_texts.append(a)
```

```python
word_list = []
for i,words in enumerate(pos_texts):
    try :
        r_A,r_B = zip(*words)
        print(i,len(r_A))
        word_list.append(r_A)
    except :
        word_list.append(" ")
word_list
```

```python
data["Text"] = word_list
```

```python
import pickle

# 저장하는 코드
with open('data.pickle', 'wb') as f:
    pickle.dump(data, f, pickle.HIGHEST_PROTOCOL)
# 불러오는 코드
with open('data.pickle', 'rb') as f:
    data = pickle.load(f) # 단 한줄씩 읽어옴
```

```python
result = []
for i in list(data["Text"]):
    result.append(" ".join(i))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf.fit(result)
```

```python
import joblib

joblib.dump(tfidf, 'tfidf.pkl')
```

```python
tfidf_save_model = joblib.load('tfidf.pkl')
```

```python
x_train_okt = tfidf_save_model.transform(result)
from lightgbm import LGBMClassifier
lgbm_clf = LGBMClassifier(n_estimators = 400, n_jobs=-1, reg_alpha = 0.9, reg_lambda = 0.2)
lgbm_clf.fit(x_train_okt, y_train)
```

```python
joblib.dump(lgbm_clf, 'sentiment_analysis.pkl')
```

```python
lgbm_clf_save_model = joblib.load('sentiment_analysis.pkl')
```

```python
comment = "귀여워❤️ 언니 올해는 5월이나 종강할때 꼭 시간내봐요 ㅎㅎ선배 넘 넘 귀엽다!!..."
content = "#하루필름"
hashtag = "하루필름"
```

```python
if comment != "" :
    comment_sentiment = list(lgbm_clf_save_model.predict(tfidf_save_model.transform([comment])))
if content != "":
    comment_sentiment.append(lgbm_clf_save_model.predict(tfidf_save_model.transform([content])))
if hashtag != "" :
    comment_sentiment.append(lgbm_clf_save_model.predict(tfidf_save_model.transform(hashtag)))

comment_sentiment_result = int(max(comment_sentiment,key = comment_sentiment.count))
comment_sentiment_result
```
